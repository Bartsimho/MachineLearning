```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

df=pd.read_csv('data/key_stage_2_ud_percentage_point_absence_attainment.csv')
df=df.drop(df[df.overall_absence_percent_groups == "Total"].index)
df=df.drop(df[df.total_number_of_pupils_at_the_end_of_key_stage_2 < 100].index)

midpoints = pd.Series([0, 2.5, 7.5, 12.5, 17.5, 22.5, 27.5, 32.5, 37.5])

plt.figure(1, figsize=(10,4))
plt.scatter(midpoints, df['percentage_reaching_the_expected_standard_in_rwm'], marker='*', c='r', label='expected level')
plt.scatter(midpoints, df['percentage_reaching_the_higher_standard_in_rwm'], c='g', label='higher level')
plt.title("Percent of Students reaching Expected RWM by Absence Percentage in KS2")
plt.xlabel("Absence Percentage")
plt.ylabel("Percent Reaching level in RWM")
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

df=pd.read_csv('data/key_stage_4_ud_percentage_point_absence_attainment.csv')
df=df.drop(df[df.overall_absence_percent_groups == "Total"].index)
df=df.drop(df[df.total_number_of_pupils_at_the_end_of_key_stage_4 < 100].index)

midpoints = pd.Series([0, 2.5, 7.5, 12.5, 17.5, 22.5, 27.5, 32.5, 37.5, 42.5, 47.5, 75])

plt.figure(1, figsize=(10,4))
plt.scatter(midpoints, df['percentage_of_pupils_who_achieved_grades_9_to_4_in_English_and_maths'], marker='*', c='r', label='Grades 9-4')
plt.scatter(midpoints, df['percentage_of_pupils_who_achieved_grades_9_to_5_in_English_and_maths'], c='g', label='Grades 9-5')
plt.scatter(midpoints, df['percentage_of_pupils_who_achieved_all_components_of_the_English_Baccalaureate_grades_9_to_5'], c='y', label='Grades 9-5 in Eng Bach')
plt.title("Percent of Students reaching Grades by Absence Percentage in KS4")
plt.xlabel("Absence Percentage")
plt.ylabel("Percent Reaching Grades")
plt.legend()
plt.show()
```
