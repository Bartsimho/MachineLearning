```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df=pd.read_csv('data/key_stage_2_ud_percentage_point_absence_attainment.csv')
df=df.drop(df[df.overall_absence_percent_groups == "Total"].index)

plt.figure(1, figsize=(10,4))
plt.scatter(df['overall_absence_percent_groups'], df['percentage_reaching_the_expected_standard_in_rwm'], marker='*', c='r', label='expected level')
plt.scatter(df['overall_absence_percent_groups'], df['percentage_reaching_the_higher_standard_in_rwm'], c='g', label='higher level')
plt.title("Percent of Students reaching Expected RWM by Absence Percentage in KS2")
plt.xlabel("Absence Percentage")
plt.ylabel("Percent Reaching level in RWM")
plt.legend()
plt.show()
```
