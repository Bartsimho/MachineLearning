```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score, mean_squared_error

df=pd.read_csv('data/key_stage_2_ud_percentage_point_absence_attainment.csv')
df=df.drop(df[df.overall_absence_percent_groups == "Total"].index)
df=df.drop(df[df.total_number_of_pupils_at_the_end_of_key_stage_2 < 100].index)

midpoints = pd.Series([0, 2.5, 7.5, 12.5, 17.5, 22.5, 27.5, 32.5, 37.5])

plt.figure(1, figsize=(10,4))
plt.scatter(midpoints, df['percentage_reaching_the_expected_standard_in_rwm'], marker='*', c='r', label='expected level')
plt.scatter(midpoints, df['percentage_reaching_the_higher_standard_in_rwm'], c='g', label='higher level')
plt.title("Percent of Students reaching level of RWM by Absence Percentage in KS2")
plt.xlabel("Absence Percentage")
plt.ylabel("Percent Reaching level in RWM")
plt.legend()

Model1 = np.poly1d(np.polyfit(midpoints, df['percentage_reaching_the_expected_standard_in_rwm'], 3))
Model2 = np.poly1d(np.polyfit(midpoints, df['percentage_reaching_the_higher_standard_in_rwm'], 3))
Line = np.linspace(0, 37.5, 9)

plt.plot(Line, Model1(Line), 'r--')
plt.plot(Line, Model2(Line), 'g--')
plt.show()

y1_pred = Model1(Line)
y2_pred = Model2(Line)

RMSE1 = np.sqrt(mean_squared_error(df['percentage_reaching_the_expected_standard_in_rwm'], y1_pred))
RMSE2 = np.sqrt(mean_squared_error(df['percentage_reaching_the_higher_standard_in_rwm'], y2_pred))

print("Expected Level R2 Score: " + str(r2_score(df['percentage_reaching_the_expected_standard_in_rwm'], Model1(midpoints))) + " and a RMSE of: " + str(RMSE1))
print("Higher Level R2 Score: " + str(r2_score(df['percentage_reaching_the_higher_standard_in_rwm'], Model2(midpoints)))+ " and a RMSE of: " + str(RMSE2))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import r2_score

df=pd.read_csv('data/key_stage_4_ud_percentage_point_absence_attainment.csv')
df=df.drop(df[df.overall_absence_percent_groups == "Total"].index)
df=df.drop(df[df.total_number_of_pupils_at_the_end_of_key_stage_4 < 100].index)

midpoints = pd.Series([0, 2.5, 7.5, 12.5, 17.5, 22.5, 27.5, 32.5, 37.5, 42.5, 47.5, 75])

plt.figure(1, figsize=(10,4))
plt.scatter(midpoints, df['percentage_of_pupils_who_achieved_grades_9_to_4_in_English_and_maths'], marker='*', c='r', label='Grades 9-4')
plt.scatter(midpoints, df['percentage_of_pupils_who_achieved_grades_9_to_5_in_English_and_maths'], c='g', label='Grades 9-5')
plt.scatter(midpoints, df['percentage_of_pupils_who_achieved_all_components_of_the_English_Baccalaureate_grades_9_to_5'], c='y', label='Grades 9-5 in Eng Bacc')
plt.title("Percent of Students reaching Grades by Absence Percentage in KS4")
plt.xlabel("Absence Percentage")
plt.ylabel("Percent Reaching Grades")
plt.legend()

Model1 = np.poly1d(np.polyfit(midpoints, df['percentage_of_pupils_who_achieved_grades_9_to_4_in_English_and_maths'], 3))
Model2 = np.poly1d(np.polyfit(midpoints, df['percentage_of_pupils_who_achieved_grades_9_to_5_in_English_and_maths'], 3))
Model3 = np.poly1d(np.polyfit(midpoints, df['percentage_of_pupils_who_achieved_all_components_of_the_English_Baccalaureate_grades_9_to_5'], 3))
Line = np.linspace(0, 75, 12)

plt.plot(Line, Model1(Line), 'r--')
plt.plot(Line, Model2(Line), 'g--')
plt.plot(Line, Model3(Line), 'y--')
plt.show()

y1_pred = Model1(Line)
y2_pred = Model2(Line)
y3_pred = Model3(Line)

RMSE1 = np.sqrt(mean_squared_error(df['percentage_of_pupils_who_achieved_grades_9_to_4_in_English_and_maths'], y1_pred))
RMSE2 = np.sqrt(mean_squared_error(df['percentage_of_pupils_who_achieved_grades_9_to_5_in_English_and_maths'], y2_pred))
RMSE3 = np.sqrt(mean_squared_error(df['percentage_of_pupils_who_achieved_all_components_of_the_English_Baccalaureate_grades_9_to_5'], y3_pred))

print("Grades 9-4 R2 Score: " + str(r2_score(df['percentage_of_pupils_who_achieved_grades_9_to_4_in_English_and_maths'], Model1(midpoints))) + " and a RMSE of: " + str(RMSE1))
print("Grades 9-5 R2 Score: " + str(r2_score(df['percentage_of_pupils_who_achieved_grades_9_to_5_in_English_and_maths'], Model2(midpoints))) + " and a RMSE of: " + str(RMSE2))
print("Grades 9-5 Eng Bacc R2 Score: " + str(r2_score(df['percentage_of_pupils_who_achieved_all_components_of_the_English_Baccalaureate_grades_9_to_5'], Model3(midpoints))) + " and a RMSE of: " + str(RMSE3))
```
